```python
data = pd.read_csv('segmentation_data.txt',sep=',', header=None)
print(data.shape)
data.head()
```

```python
class_lab=pd.read_csv('segmentation_classes.txt',sep='\t', header=None)
print(class_lab.shape)
class_lab.head()
```

```python
print(class_lab.isnull().sum())
print(data.isnull().sum())
```

```python
np.set_printoptions(precision=4,suppress=True)
#floating point numbers using fixed point notation
```

```python
seg_data=data.values
```

```python
#perform min-max normalization on the data matrix
scaler = MinMaxScaler()
print(scaler.fit(seg_data))
print(scaler.data_max_)
```

```python
normalize_data=scaler.transform(seg_data)
print(normalize_data)
print(normalize_data.shape)
```

```python
label=class_lab[1].to_numpy()
label.shape
```

```python
#b.perform kmeans clustering
kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
```

```python
kmeans.fit(normalize_data)
```

```python
seg_names=pd.read_csv('segmentation_names.txt',sep='\t', header=None)
print(seg_names.values.tolist())
```

```python
clusters = kmeans.predict(normalize_data)
print(clusters)
print(kmeans.labels_)
print(seg_names.values.tolist())
print(kmeans.cluster_centers_)
```

```python
from sklearn.metrics import completeness_score, homogeneity_score
```

```python
print('completeness_score = ' + str(completeness_score(label,clusters)))
print('homogeneity_score = '+str(homogeneity_score(label,clusters)))
```

```python
#c. PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
seg_pca = pca.fit_transform(normalize_data)
print(seg_pca)
```

```python
print(pca.explained_variance_ratio_)   # 4 components can capture 95.46% fo variancne
```

```python
pca_4 = PCA(n_components=4) # using 4 components 
seg_pca4 = pca_4.fit_transform(normalize_data) #transform data into a reduce dimension
print(seg_pca4)
```

```python
#d
kmeans.fit(seg_pca4)
```

```python
c_pca = kmeans.predict(seg_pca4)
print(c_pca)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
print('completeness_score = ' + str(completeness_score(label,c_pca)))
print('homogeneity_score = '+str(homogeneity_score(label,c_pca)))
```
