```python
bank_df_numeric = bank_df[["age","income","children"]]
bank_df_norm = (bank_df_numeric - bank_df_numeric.min()) / (bank_df_numeric.max()-bank_df_numeric.min())
bank_df_norm .head(10)
```

```python
zscore = lambda x: (x - x.mean()) / x.std()
bank_numeric = bank_df[["income"]]
bank_num_std = bank_numeric.apply(zscore)
bank_num_std.head(10)
```

```python
#4.Discretize the age attribute into 3 categories (corresponding to "young", "mid-age", and "old").
inc_bins = pd.qcut(bank_df["age"], [0, .33, .66, 1], labels=["young", "mid-age", "old"], retbins=True)
inc_bins
```

```python
df_dum=pd.get_dummies(bank_df, prefix=['gender', 'married','car','savings_acct','region','current_acct','mortgage','pep'],columns=['gender', 'married','car','savings_acct','region','current_acct','mortgage','pep'])
df_dumm=df_dum.drop(columns=['age','income','children'])
df_dumm.head()
```

```python
result = pd.concat([df_dumm,bank_df_norm], axis=1)
result.head()
```

```python
result.columns
```

```python
res=result[['id','age', 'income', 'children', 'gender_FEMALE', 'gender_MALE', 'married_NO', 'married_YES',
       'car_NO', 'car_YES', 'savings_acct_NO', 'savings_acct_YES',
       'region_INNER_CITY', 'region_RURAL', 'region_SUBURBAN', 'region_TOWN',
       'current_acct_NO', 'current_acct_YES', 'mortgage_NO', 'mortgage_YES',
       'pep_NO', 'pep_YES']]
res.head()
```

```python
res.to_csv("bank_numeric.csv", float_format="%1.2f")
```
