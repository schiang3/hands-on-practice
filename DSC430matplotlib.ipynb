```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def conway(s, p):
    'a board, is a square two dimensional NumPy array [0,1] of size s'
    return np.random.choice([0,1],(s,s),p=[1-p,p])
```

```python
def ruleAdvance(b,s):
    'rule of live cell and one time broad'
    for i in range(s):
        for j in range(s):
            #count neighbors up,down,left,right,up-right,up-left,left-down,right-down
            total=int(b[i,(j+1)%s]+b[i,(j-1)%s]+b[(i+1)%s,j]+b[(i-1)%s,j]+b[(i-1)%s,(j-1)%s]+b[(i-1)%s,(j+1)%s]+b[(i+1)%s,(j-1)%s]+b[(i+1)%s,(j+1)%s])
            if b[i,j]==1:#live cell
                if total>3:#die
                    b[i,j]=0#die
                elif total<2:#die
                    b[i,j]=0#die
            else: #dead cell
                if total==3: #total=3 make cell alive
                    b[i,j]=1
    return b
```

```python
def advance(b,t):
    'accepts a Conway board and advances it t time steps' 
    s=b.shape[0]
    plt.imshow(b)
    plt.show()
    for t in range(t):
        b=ruleAdvance(b,s)
        plt.imshow(b)
        plt.show()
        print('advanced one more time')
```

```python
def display(s,p,t):
    b=conway(s,p)
    advance(b,t)
    
```
